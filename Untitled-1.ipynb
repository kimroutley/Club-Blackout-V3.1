# Local Audit Runner

Run these from the project root (where `pubspec.yaml` lives):

- Static analysis: `flutter analyze`
- Unit/widget tests: `flutter test -r expanded`
- If you use integration tests: `flutter test integration_test -r expanded`

If anything fails, paste the first failing stack trace + the test name.

```python
# Optional helper: paste `flutter test -r expanded` output into a file and grep for failures.
import re, pathlib

p = pathlib.Path('test_output.txt')
if p.exists():
    text = p.read_text(encoding='utf-8', errors='ignore')
    m = re.search(r'^(.*?FAILED.*?)$', text, flags=re.MULTILINE)
    print(m.group(1) if m else 'No FAILED line found.')
else:
    print('Create test_output.txt in this folder with flutter test output.')
```
