# Simulation Data Accumulation Setup
The best way to accumulate accurate data for the 'Win Odds' predictor is a **Monte Carlo Simulation**. 
By running thousands of automated games based on different starting role distributions, we can build a baseline "Win Probability Matrix."

Below is the logic for a custom script you can run (typically in a dev environment or a standalone dart runner) to gather this data.


```python
# Note: This is pseudo-code for a Dart Simulation Script 
# You would run this using 'dart run' on a file in your project.

"""
import 'package:club_blackout/logic/game_engine.dart';

void runSimulations(int iterations) async {
  Map<String, int> winCounts = {};
  
  for (int i = 0; i < iterations; i++) {
    // 1. Setup a fresh engine
    final engine = GameEngine(roleRepo: myRepo);
    
    // 2. Randomize a game of size X
    await engine.createTestGame(roleCount: 12); 
    await engine.startGame();
    
    // 3. Automated Playloop
    while (!engine.isGameOver) {
      if (engine.currentPhase == GamePhase.night) {
        // Auto-perform random valid actions for the script
        engine.performRandomNightActions();
      } else {
        // Auto-perform random voting
        engine.performRandomDayVote();
      }
      engine.nextPhase();
    }
    
    // 4. Collect results
    String winner = engine.winner ?? 'DRAW';
    winCounts[winner] = (winCounts[winner] ?? 0) + 1;
  }
  
  print("Simulation Complete: $winCounts");
}
"""
```
